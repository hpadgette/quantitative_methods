### Part 1: Looping and control flow

a) Print all the numbers in the Fibonacci sequence less than $10^{6}$.

````python
prev_num = 0 # fibonacci sequence starts at 0
next_num = 1 # next number in the sequence is 1

print(prev_num) # prints the first number in the sequence
print(next_num) # prints the next number in the sequence

next_next_num = 0 # just delcares the next_next_num variable before the loop so can end the loop dependening on the value of next_next_num
max_num = 10**6   # prints all the numbers in the Fibonacci sequence up to this number

while next_next_num < max_num:                   # iterates through the loop while the next fibonacci sequence value is less than the value we want to print up until
    next_next_num = prev_num + next_num          # this is what defines the fibonacci sequence - the next number is the sum of the previous two
    if next_next_num > max_num: break            # this introduces a way to break the loop if the next number is greater than our max desired value (otherwise we'd print one more number in the sequence than we really want)
    print(next_next_num)                         # prints the numbers in the Fibonacci sequence!
    prev_num, next_num = next_num, next_next_num # changes next_num to be pre_num and next_next_num to be next_num so that we keep building and printing the Fibonacci sequence
````

a) How many times does the word “will” appear? Try to do a count that catches capitalized and uncapitalized versions, that doesn’t include non-Shakepear text like copyright notices (without manually editing the file to remove them). Make sure you don’t count the characters “will” when it appears in the middle of another word.

b) How many different words are used in the complete works of William Shakespeare? Make sure you strip out punctuation, and that words with both upper and lowercase are counted as the same word. Don’t include numbers, and don’t include the nontextual information about the text. 

c) Print out a list of the 30 most common words in the complete works of William Shakespeare.

Some functions and methods that may be useful: `open()`, `readlines()`, `close()`, and the different methods for the data type string (Google “python string methods“). Google is your friend with trying to figure out things in Python! Most questions that possibly could be answered have been answered.

````python
###### PART A ######

# open the text file and create a list shaks12_lines containing all lines in the text file as strings
with open("shaks12.txt", 'r') as shaks12:
    shaks12_lines = shaks12.readlines()

sep = [] # initializing the sep list

# iterates through all lines
# removes the \n characters at the end of each line 
for line in shaks12_lines:
    sep.append(line.strip())

keyword = "will" # sets the word that we're searching for. in this problem, it's "will"
count = 0        # initialize the count variable and set it equal to zero

for i in sep:
    # this is where the "re" module imported above is used - for each line separates all text from all punctuation 
    i = re.findall(r'\w+|[^\w\s]', i)
    
    for j in i:                    # iterates through all words in all lines (now words and punctuation separated)
        j = j.lower()              # makes all words in all lines lowercase (so not case-sensitive)
        if j == keyword: count +=1 # counts all instances of the keyword

print("The word " + keyword + " appears " + str(count) + " times in this text file.") # prints the result
````

````python
###### PART B ######

test_lines = open("shaks12.txt", 'r')          # opens the file
cut_lines = test_lines.readlines()[253:124437] # only reads the lines after the intro copyright info and before the end copy right info
                                               # stores to the cut_lines list

# this section removes all instances of the copyright notice that appears within the bulk text.
# since it's the same 8 lines every time, I made strings of all those lines, iterated through all those lines in the text, 
# and then removed all instances of that line from the list of lines using the .pop(i) function discussed in class on 30 Aug 23

target_string1 = "<<THIS ELECTRONIC VERSION OF THE COMPLETE WORKS OF WILLIAM\n"
target_string2 = "SHAKESPEARE IS COPYRIGHT 1990-1993 BY WORLD LIBRARY, INC., AND IS\n"
target_string3 = "PROVIDED BY PROJECT GUTENBERG ETEXT OF ILLINOIS BENEDICTINE COLLEGE\n"
target_string4 = "WITH PERMISSION.  ELECTRONIC AND MACHINE READABLE COPIES MAY BE\n"
target_string5 = "DISTRIBUTED SO LONG AS SUCH COPIES (1) ARE FOR YOUR OR OTHERS\n"
target_string6 = "PERSONAL USE ONLY, AND (2) ARE NOT DISTRIBUTED OR USED\n"
target_string7 = "COMMERCIALLY.  PROHIBITED COMMERCIAL DISTRIBUTION INCLUDES BY ANY\n"
target_string8 = "SERVICE THAT CHARGES FOR DOWNLOAD TIME OR FOR MEMBERSHIP.>>\n"

for line in range(len(cut_lines) - 1, -1, -1):
    if cut_lines[line] == target_string1: cut_lines.pop(line)
    if cut_lines[line] == target_string2: cut_lines.pop(line)
    if cut_lines[line] == target_string3: cut_lines.pop(line)
    if cut_lines[line] == target_string4: cut_lines.pop(line)
    if cut_lines[line] == target_string5: cut_lines.pop(line)
    if cut_lines[line] == target_string6: cut_lines.pop(line)
    if cut_lines[line] == target_string7: cut_lines.pop(line)
    if cut_lines[line] == target_string8: cut_lines.pop(line)
    
words_dict = dict() # create an empty dictionary that distinct words and their appearances can be counted

for line in cut_lines:
    line = line.lower()     # makes all words lower-case so no case sensitivity (e.g., "Will" and "will" are counted as just one word
    line = line.strip()     # removes whitespace / new line characters
    line = line.translate(line.maketrans("", "", string.punctuation)) # removes all punctuation characters
    words = line.split(" ") # separates out individual words in each line of text

    for word in words: # iterates through all the words in the text 
        if word in words_dict: # if a word is already in the dictionary, add 1 to its count
            words_dict[word] = words_dict[word] + 1
        else: # if a word is not yet in the dictionary, add it to the dictionary and make its count 1
            words_dict[word] = 1

del words_dict[''] # deletes blank lines count from dictionary

# sorts the dictionary from highest word count to lowest word count
sorted_dict = dict(sorted(words_dict.items(), key=lambda item: item[1], reverse=True))

# creates a list of keys to delete (all numerical digits) 
delete = []

# iterates through all keys in the dictionary
# if a key has a numerical digit in it, add that key to the delete dicitionary
for key in sorted_dict:
    if '0' in key or '1' in key or '2' in key or '3' in key or '4' in key or '5' in key or '6' in key or '7' in key or '8' in key or '9' in key:
        delete.append(key)

# for every key in the keys_to_delete list, delete it from the sorted dictionary
# this gets around the error which is thrown when trying to delete a key while iterating through a list / dictionary
for key in delete:
    del sorted_dict[key]

# initialize the count variable and set its value to 0
count = 0 

# iterates through all the keys in the sorted dictionary and counts how many there are
for key in list(sorted_dict.keys()):
    count += 1

# prints the result
print("There are " + str(count) + " words in the complete works of William Shakespeare.")
````

- Now, transitioning to seaborn's tutorial on high-level API for statistical graphics.
- relplot(), which is used above, can visualize different kinds of statistical relationships. This includes the scatter plot generated above, but this is not always an effective way to graph data. For instance, relationships wherein the x axis is time might be better represented by a line. The relplot() ```kind``` parameter enables the user to switch to alternative representations. 

````python
dots = sns.load_dataset("dots")

# plotting the data WITHOUT specifying a line plot
sns.relplot(
    data=dots, 
    x="time", y="firing_rate", col="align",
    hue="choice", size="coherence", style="choice",
    facet_kws=dict(sharex=False),
)

# plotting the data WITH specifying a line plot
sns.relplot(
    data=dots, kind="line",
    x="time", y="firing_rate", col="align",
    hue="choice", size="coherence", style="choice",
    facet_kws=dict(sharex=False),
)
````

- The two codes graphed above are of exactly the same data but with one additional parameter passed to relplot(), we can generate a plot which better visualizes the data we're trying to analyze.
- The size and size parameters change the marker area and symbol in the scatter plot but the line width and dashing in the line plot.
